```python
from aocd import get_data, submit
```

```python
input = get_data(day=1, year=2025)
```

```python
part_1_input = input.split()
```

```python
def solution(codes: list[str])->int:
    ans = 0
    val = 50
    for code in codes:
        dir = -1 + 2 * (code[0] == "R")
        magnitude = int(code[1:])
        val = (val + dir * magnitude) % 100

        if val == 0:
            ans += 1
    return ans
```

```python
solution([
    "L68",
    "L30",
    "R48",
    "L5",
    "R60",
    "L55",
    "L1",
    "L99",
    "R14",
    "L82",
])
```

```python
solution(part_1_input)
```
