```python
def safe_is_paper(x:int, y:int, arr: list[list[str]]):
    try:
        if y < 0 or x < 0:
            return 0 
        if arr[y][x] == '@':
            return 1 
    except IndexError:
        return 0
    return 0
```

```python
def solution(input):
    ans = 0
    for x in range(len(input[0])):
        for y in range(len(input)):
            if input[y][x] == '@':
                adj = [safe_is_paper(x+j,y+i, input) for i in [-1,0,1] for j in [-1,0,1]]
                if sum(adj) < 5:
                    ans += 1
    return ans
```
