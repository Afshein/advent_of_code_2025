```python
from aocd import get_data, submit
```

```python
input = get_data(day=3, year=2025)
```

```python
batteries = input.split()
```

```python
def solution(batteries: list[int]) -> int:
    ans = 0
    for battery in batteries:
        running_max = [0]
        for num in battery[::-1]:
            num = int(num)
            if num > running_max[-1]:
                running_max.append(num)
        
        joltage = running_max[-1] * 10 + running_max[-2]
        ans += joltage
    return ans
```

```python
solution(batteries)
```
